```python
import os
import sys
import csv
import matplotlib.pyplot as plt
import numpy as np
sys.path.append('/home/vincent/Documents/Urbanhack2014/urbandatahack2014')
import urbandata.parser as parser
import urbandata.kde as kde
from scipy import stats
import pandas as pd
filename = './../data/WCC_CleansingAntiSocialBehaviour.csv'

# Loading the data
data_anti_social = parser.get_anti_social_data()
data_pub = parser.get_licensing_data(True)

#data_vomit = data_anti_social[data_anti_social['Vomit']]

#data_pub = data_licensing[data_licensing['Premisesuse'] == 'Type - Night clubs and discos']
def coord_from_data(data):
    x = -np.asarray(data['lon_to_gw'])
    y = np.asarray(data['lat_to_gw'])
    I = np.where(x != 0)[0]
    x =x[I]
    y =y[I]
    return (x,y)
```

```python
# density estimation on pubs
x_p = np.asarray(data_pub['Long'])
y_p = np.asarray(data_pub['Lat'])
I = np.where(x_p != 0)[0]
x_p =x_p[I]
y_p =y_p[I]

(x_p,y_p) = coord_from_data(data_pub)

k_p =  kde.get_kernel_density(x_p,y_p)
Z_p = kde.evaluate_kernel_density_on_grid(x_p,y_p,k_p)

```

```python
# density estimation on all antisocial behavior
data_as = data_anti_social
(x_as,y_as) = coord_from_data(data_as)

k_as =  kde.get_kernel_density(x_as,y_as)
Z_as = kde.evaluate_kernel_density_on_grid(x_as,y_as,k_as)
```

```python
kde.plot_kde(x_p,y_p,Z_p)
kde.plot_kde(x_as,y_as,Z_as)
```

```python

xmin = x_as.min()
xmax = x_as.max()
ymin = y_as.min()
ymax = y_as.max()
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.plot(x_p, y_p, 'r.', markersize=5)
ax.plot(x_as, y_as, 'k.', markersize=5)
ax.axis('equal')
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])

plt.ylabel('(m)')
plt.xlabel('(m)')

plt.show()
```

```python
# distance to nearest pub
import scipy.spatial.distance as dist
X_p =np.transpose(np.vstack([x_p,y_p]))
X_as =np.transpose(np.vstack([x_as,y_as]))
d = dist.cdist(X_p,X_as)
d_min = np.amin(d,axis=0)
```

```python
# evaluating pub density on anti social behavior points


data_as = data_anti_social[data_anti_social['Urine']]
(x_as,y_as) = coord_from_data(data_as)


Z_as = kde.evaluate_kernel_density_on_grid(x_as,y_as,k_p)
#kde.plot_kde(x,y,Z)

k_as =  kde.get_kernel_density(x_as,y_as)

nbins = 50

z_as_p = kde.evaluate_kernel_density(x_as,y_as,k_p)
z_as_as = kde.evaluate_kernel_density(x_as,y_as,k_as)

plt.plot(z_as_p,z_as_as,'*')
plt.xlabel('antisocial density')
plt.ylabel('pub density')
plt.show()

plt.hist(z_as_p,nbins)
plt.ylabel('antisocial count')
plt.xlabel('pub density')
plt.show()

plt.hist(log(z_as_p),nbins)
plt.ylabel('antisocial count')
plt.xlabel('log pub density')
plt.show()


X_as =np.transpose(np.vstack([x_as,y_as]))
d = dist.cdist(X_p,X_as)
d_min = np.amin(d,axis=0)
plt.hist(d_min,nbins)
plt.ylabel('antisocial count')
plt.xlabel('distance to nearest pub density')
plt.show()


plt.plot(1/d_min,z_as_as,'*')
plt.ylabel('antisocial count')
plt.xlabel('inverse distance to nearest pub')
plt.show()
```
